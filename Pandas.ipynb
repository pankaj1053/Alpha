```python
#Merging, Concatenating and Appending
df1=pd.DataFrame({'id': [1,2,3,4,5],
                 'name':['a','b','c','d','e'],
                 'sub':['sub1','sub2','sub3','sub4','sub5']
})
df2=pd.DataFrame({
    'id': [1,2,3,4,5],
                 'name':['b','c','d','e','f'],
                 'sub':['sub3','sub4','sub5','sub6','sub7']
    
})
```

```python
#Concatenating two DataFrames
pd.concat([df1,df2],axis=0)         #joining two DataFrames along the rows
```

```python
pd.concat([df1,df2],axis=1)      #Joining 2 DataFrames loang the columns
#These methods are widely used in Feature Engineering
```

```python
#Merging Two DataFrames
pd.merge(left=df1,right=df2,on='sub')     #joining 2 Dataframes using sub as key
```

```python
pd.merge(left=df1,right=df2,on='id')
```

```python
#Multi -Index DataFrame
Company=['Google','Google','Google','Microsoft','Microsoft','Microsoft']
Year=[2008,2009,2010,2008,2009,2010]
Revenue=[11,15,16,9,12,14]
Employee=[300,400,500,350,450,550]
```

```python
list(zip(Revenue,Employee))      #Zip will collect one value from each of its container(Revenue,Employee)
```

```python
list(zip(Company,Year))
```

```python
hier_index=list(zip(Company,Year))           #these pair values will be our 2 indices
```

```python
hier_index=pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
multi_index_df=pd.DataFrame(data=list(zip(Revenue,Employee)),index=hier_index, columns=['Revenue','Employee'])
multi_index_df
```

```python
multi_index_df.index.names=['Company','Year']
#rename our indices to Company  and Year
multi_index_df
```

```python
multi_index_df.loc['Google']
#Accessing data using our first index
```

```python
multi_index_df.loc['Google'].loc[2009]
#Accessing Data using 1st index and 2nd index
```

```python
multi_index_df.xs('Microsoft',level='Company')
#Accessing data based on Index level(either from first or second index)
```

```python
multi_index_df.xs(2010,level='Year')
#Accessinf data based on Index level
```
